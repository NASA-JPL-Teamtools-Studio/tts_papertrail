```python
from datetime import datetime
from tts_data_utils.core.generic import GenericContainer
from tts_papertrail.base import Report, TextEntry, ListEntry, ListItem, SectionEntry, SummaryTable, PowerTableEntry, RichText
```

# Papertrail Demonstration

Papertrail is a library meant to be a centralized reporting tool. It works via a single `Report` class that allows users to build a report piece by piece, and then write it as html, powerpoint, work, or excel.

This should allow developers to work in the same way regardless of the particular requirements of their projects.

We begin by creating some dummy data with the lorem feature of DatContainer just so we can make some tables below

```python
mission_data = GenericContainer(name="Mission Telemetry", lorem=5)
system_status = GenericContainer(name="Subsystem Health", lorem=3)
```

## Initialize the report

```python
report = Report(name="JPL Mission Status Report", author="Operations Team")
```

## Adding a Section

```python
intro_text = TextEntry(
    title="Executive Summary",
    content=[
        RichText("Classification: ", bold=True),
        RichText("Internal Use Only", color="#FF0000"),
        "\nThis report summarizes mission progress and telemetry health."
    ]
)
report.add_entry(intro_text)
```

## Summary Table

The summary table is a sepcial entry that allows you to add rows as you go. The idea is that you will initialize the summary table in line where you'd like it to go, and then call the `add_to_summary_table` method as you go. 

1. Add summary table
2. Add new section
3. Use `add_to_summary_table` to add a summary of what you just added
4. Repeat rows 2 and 3 until done.

```python
summary = SummaryTable(title="Deployment Status Overview")
report.add_entry(summary)
```

## Filling out some more sections

```python
report.add_entry(PowerTableEntry(title="Telemetry Log", container=mission_data))
summary.add_to_summary_table("Telemtry Log", "PASS", "All sensors within limits", "Nominal")

report.add_entry(PowerTableEntry(title="Hardware Checklist", container=system_status))
summary.add_to_summary_table("Hardware Checklist", "WARNING", "Solar array 2 tracking lag", "Monitor")

nested_list = ListEntry(
    title="Next Steps",
    items=[
        ListItem(level=0, content=[RichText("Phase 1: Deployment", bold=True)]),
        ListItem(level=1, content=[RichText("Calibrate primary sensors")]),
        ListItem(level=1, content=[RichText("Verify downlink bandwidth")]),
        ListItem(level=0, content=[RichText("Phase 2: Science Ops", italic=True)])
    ]
)

summary.add_to_summary_table("Next Steps", "PASS", "Ready for next steps", "Nominal")
report.add_entry(nested_list)
```

## Generate Report in Each fOrmat

```python
report.to_html("mission_report.html")        # Interactive tabbed dashboard
report.to_excel("mission_report.xlsx")      # Spreadsheet with frozen headers
report.to_word("mission_report.docx")       # Formatted document with page breaks
report.to_powerpoint("mission_report.pptx") # Presentation with paginated tables
```
